```python
def dataFrame_creation(file_Name):
  data = []
  with open(file_Name, 'r') as file:
      for line in file:
          values = line.strip().split(':::')  # Use correct delimiter
          data.append(values)
  return data
import pandas as pd

# Create DataFrame with headers
column_names = ['ID', 'Title', 'Genre', 'Description']
col = ['ID','Title','Description']
df_train = pd.DataFrame(dataFrame_creation('train_data.txt'), columns=column_names)
print(df_train.head())
df_test = pd.DataFrame(dataFrame_creation('test_data.txt'), columns = col)
```

```python
df_train[['Title','Year']] = df_train['Title'].str.split('(', expand = True, n = 1)
df_train['Year'] = df_train['Year'].str.rstrip(' )')
df_train.head(n=10)
df_train.tail(n=10)
```

```python
#Data Preprocessing
# Lowercasing
df_train['Description'] = df_train['Description'].str.lower()
df_train.head()
```

```python
#Removing Punctuation
import re
df_train['preprocessed_Des'] = df_train['Description'].str.replace(r"[#$%&!@'\d+]","")
df_train.head()
```

```python
#tokenization, Lemmatization
import nltk
nltk.download('punkt_tab')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
df_train['tokens'] = df_train['preprocessed_Des'].apply(word_tokenize)
```

```python
#Loading and removing stop words
stop_words = set(stopwords.words('english'))
df_train['tokens'] = df_train['tokens'].apply(lambda words:[word for word in words if word not in stop_words])
df_train.head()
```

```python
#Lemmatization
nltk.download('wordnet')
lematizer = WordNetLemmatizer()
df_train['tokens'] = df_train['tokens'].apply(lambda words : [lematizer.lemmatize(word) for word in words])
df_train.head()
```

```python
#Label encoding the genre column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(df_train['Genre'])
```

```python
#Text Vectorization
df_train['tokens'] = df_train['tokens'].apply(lambda x: ",".join(x))
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features = 5000, ngram_range =(1,2), max_df = 0.8, min_df = 5)
X = vectorizer.fit_transform(df_train['tokens'])
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 123)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
```

```python
#Model metrics evaluation
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred,target_names = le.classes_))
print(accuracy_score(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, 'genre_classification.pkl')

joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```

```python
#Loading the model
loaded_model = joblib.load('genre_classification.pkl')

loaded_vector = joblib.load('tfidf_vectorizer.pkl')
```

```python
#Using on the test data
print(df_test)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    cleaned_text = ' '.join(tokens)
    return cleaned_text

df_test['Description'] = df_test['Description'].apply(preprocess)
```

```python
X_new = loaded_vector.transform(df_test['predicted'])
```

```python
predictions = loaded_model.predict(X_new)
```

```python
joblib.dump(le, 'label_encoder.pkl')
label_encoder = joblib.load('label_encoder.pkl')
```

```python
#Decoding the predictions
df_test['Predicted_genres'] = label_encoder.inverse_transform(predictions)
print(df_test.head())
df_test.to_csv('Results.csv')
```
